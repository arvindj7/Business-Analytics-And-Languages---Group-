```python
df_summer = df[df['Season'] == 'Summer'] # Filter for the Summer Olympics
df_summer['Host NOC'] = df_summer['City'].map(city_to_noc) # Adding a column mapping host cities to NOC codes
df['Host NOC'] = df['City'].map(city_to_noc) # Droping rows with missing values 
df = df.dropna(subset=['City', 'NOC', 'Medal']) # Droping rows with missing values 

medals_by_country = df.groupby(['Year', 'NOC'])['Medal'].count().reset_index() # Calculate total medals won by each country in each year
medals_by_country.columns = ['Year', 'NOC', 'Total Medals']

hosts = df[['Year', 'Host NOC']].drop_duplicates() # Identify host performance by merging with host NOC
host_performance = medals_by_country.merge(hosts, left_on='Year', right_on='Year')
host_performance = host_performance[host_performance['NOC'] == host_performance['Host NOC']]

# Determine if the host won the most medals in their hosting year
# Compare host performance with all other nations in the same year
host_performance['Did Host Win Most Medals'] = host_performance.apply(
    lambda row: row['Total Medals'] == medals_by_country[
        medals_by_country['Year'] == row['Year']
    ]['Total Medals'].max(),
    axis=1
)

# Display results
(host_performance[['Year', 'Host NOC', 'Total Medals', 'Did Host Win Most Medals']])
```

```python
# Calculate total medals won by each country in each year
medal_counts = df_summer.groupby(['Year', 'NOC'])['Medal'].count().reset_index()
medal_counts.columns = ['Year', 'NOC', 'Total Medals']

# Rank countries by total medals in each year
medal_counts['Rank'] = medal_counts.groupby('Year')['Total Medals'].rank(ascending=False, method='min')

# Filtering only host countries and explicitly create a copy
host_medals = medal_counts[medal_counts['NOC'].isin(df_summer['Host NOC'])].copy()

# Checking if host countries were in the top
host_medals.loc[:, 'Top'] = host_medals['Rank'] <= 1

# Merge host performance with the host years and NOCs
host_results = host_medals.merge(
    df_summer[['Year', 'Host NOC']].drop_duplicates(),
    left_on=['Year', 'NOC'],
    right_on=['Year', 'Host NOC']
)

# Focus on relevant columns
host_results = host_results[['Year', 'Host NOC', 'Total Medals', 'Rank', 'Top']]

# Sort results by Year
host_results = host_results.sort_values(by='Year')

# Visualization: Host country medal performance over time
plt.figure(figsize=(14, 7))
sns.barplot(data=host_results, x='Year', y='Total Medals', hue='Top', dodge=False)

# Customize the plot
plt.title('Host Country Medal Performance (Summer Olympics)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Total Medals', fontsize=14)
plt.xticks(rotation=45, fontsize=10)
plt.legend(loc='upper right')
plt.tight_layout()
plt.savefig('host_country_performance.png', format = "png", dpi=300, bbox_inches='tight')  # Export as PNG
plt.show()
```

This graph presents a comparison of the performance of host nations at the Summer Olympics across the years. From this graph, it becomes quite clear that the medal count for the host country is substantially higher during the year it hosts the Games, which upholds the well-accepted "home advantage" effect. Hosting has a number of advantages, including familiarity with local conditions, a sea of crowd support, and higher investments in sports infrastructure and training (Rose & Spiegel, 2011). This is a well-noticeable trend over the years, with countries like the United States and China performing exceptionally well in 1984 and 2008, respectively. The difference in medal count each year the country has hosted is representative not just of the fluctuating investment by the nations but also of the difference in their sporting capabilities at that time. This graph forms a very telling visual of how hosting the Olympics may serve as a catalyst to succeed.

##### Participation of Females Over the years

```python
# Filtering out rows with missing gender information
df = df.dropna(subset=['Sex'])

# Filtering only Summer Olympics 
df = df[df['Season'] == 'Summer']

# Total number of athletes per year
total_athletes = df.groupby('Year')['ID'].nunique().reset_index()
total_athletes.columns = ['Year', 'Total Athletes']

# Number of female athletes per year
female_athletes = df[df['Sex'] == 'F'].groupby('Year')['ID'].nunique().reset_index()
female_athletes.columns = ['Year', 'Female Athletes']
female_athletes
```

```python
# Merge total and female athlete data
participation = pd.merge(total_athletes, female_athletes, on='Year', how='left')

# Fill missing years with zero female athletes (if no data exists for early years)
participation['Female Athletes'] = participation['Female Athletes'].fillna(0)



# Plot absolute numbers of female athletes
plt.figure(figsize=(14, 7))
sns.barplot(data=participation, x='Year', y='Female Athletes', color='skyblue', alpha=0.7)
plt.title('Number of Female Athletes Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Female Athletes')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.savefig('female_participation_over_years.png', format = "png", dpi=300, bbox_inches='tight')  # Export as PNG
plt.show()
```

This graph showing the number of female athletes who have participated in the Olympics over the years vividly illustrates the increasing involvement of women in competitive sports. Early in the history of the Olympics, the participation of females was either nil or minimal, reflecting the gender roles and societal restrictions of the time. However, over the years-especially during the last decades of the 20th century-the number of female athletes increased steadily (Mitchell, 1977).

This corresponds to general social and cultural changes, including the women's rights movement, changes in politics regarding gender equality, and changes which the IOC tried to implement for greater inclusion. Other milestones that have gone a long way in enforcing such change include the introduction of new women-specific events and the necessity for equal team participation between genders. The upward trend of the data underlines progress toward gender parity in sport, while it reflects the journey and struggle involved in overcoming historical barriers.

This can, additionally, be explained by the development and support of female athletes in different countries and changes in social attitudes that encouraged women's participation in athletics. In a nutshell, the plot is a testimony to the strides made in creating an increasingly inclusive Olympic movement while reminding us of historical disparities that have gradually been addressed.

The Olympics have come a long way down the road of transformation, mapping the changes in sporting hegemony, reflecting the geopolitical realities of a period, and striving for inclusivity. It has turned into an event that not only celebrates athletic performances but also the diversity, equality, and unity of nations. These analyses show the multifaceted nature of the Olympics and their deep impact on both sports and society at large.
